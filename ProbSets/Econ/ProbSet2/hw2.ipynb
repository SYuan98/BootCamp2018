```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
from scipy.stats import norm
import matplotlib.mlab as mlab
class KDE:
    def __init__(self, X, h):
        self.data=np.sort(X)
        self.grid = np.linspace(min(X), max(X), 300)
        self.n=len(X)
        self.std=np.std(X)
        h=1.06*self.std*pow(self.n,-.2)
        self.bandwidth=h
    def kernel(self,x):
        t = 0
        for i in range(self.n):
            t = t + norm.pdf((x - self.data[i])/self.bandwidth)
        t = t / self.bandwidth / self.n
        return t
    def f(self):
        F=np.zeros(300)
        for i in range (300):
            F[i]=self.kernel(self.grid[i])
        return F
    def plot(self):
        plt.plot(self.grid,self.f(),label="Estimator")
        plt.hist(self.data, normed = True, label = "Sample distribution")
        
        plt.show()
                
            
    
```

```python
x = np.random.beta(5,5,100)
Y=KDE(x,0)
Y.plot()
```

```python
x = np.random.beta(2,5,100)
Y=KDE(x,0)
Y.plot()
```

```python
x = np.random.beta(.5,.5,100)
Y=KDE(x,0)
Y.plot()
```
