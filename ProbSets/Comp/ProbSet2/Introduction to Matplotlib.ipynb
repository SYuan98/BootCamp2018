```python
# Problem 1
import numpy as np
from matplotlib import pyplot as plt
def f(n):
    a= np.random.normal(size=(n,n))
    mean=np.zeros(n)
    mean=np.mean(a,axis=1)
    return np.var(mean)
    
```

```python
f(2)
```

```python
def g():
    b=np.zeros(10)
    for i in range (10):
        b[i]=f(100+100*i)
    plt.plot(b)
    plt.show()
```

```python
g()
```

```python
# Problem 2
def f():
    a=np.linspace(-2*np.pi,2*np.pi,100)
    y1=np.sin(a)
    y2=np.cos(a)
    y3=np.arctan(a)
    plt.plot(a,y1)
    plt.plot(a,y2)
    plt.plot(a,y3)
    plt.show()
```

```python
# Problem 3
def f():
    x1=np.linspace(-2,0.9999999,100)
    x2=np.linspace(1.0000000001,6,100)
    y1=1/(x1-1)
    y2=1/(x2-1)
    plt.plot(x1,y1,'g--',linewidth=4)
    plt.plot(x2,y2,'g--',lw=4)
    plt.xlim(-2,6)
    plt.ylim(-6,6)
    plt.show()
```

```python
f()
```

```python
# Problem 4
def f():
    x=np.linspace(0,2*np.pi,100)
    ax1=plt.subplot(221)
    ax1.plot(x,np.sin(x),'g-')
    plt.title("sin(x)",fontsize=18)
    ax2=plt.subplot(222)
    ax2.plot(x,np.sin(2*x),'r--')
    plt.title("sin(2*x)",fontsize=18)
    ax3=plt.subplot(223)
    ax3.plot(x,2*np.sin(x),'b--')
    plt.title("2*sin(x)",fontsize=18)
    ax4=plt.subplot(224)
    ax4.plot(x,2*np.sin(2*x),'c-.')
    plt.title("2*sin(2*x)",fontsize=18)
    plt.axis([0,2*np.pi,-2,2])
    plt.suptitle(" figure")
```
