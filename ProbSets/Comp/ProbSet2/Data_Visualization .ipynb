```python
# Problem 1
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd 
from scipy import stats 
d=np.load("anscombe.npy")
df1=pd.DataFrame(d[:,:2],columns=['x','y'])
df2=pd.DataFrame(d[:,2:4],columns=['x','y'])
df3=pd.DataFrame(d[:,4:6],columns=['x','y'])
df4=pd.DataFrame(d[:,6:],columns=['x','y'])
ax1=plt.subplot(221)
ax2=plt.subplot(222)
ax3=plt.subplot(223)
ax4=plt.subplot(224)
df1.plot(kind="scatter",x="x",y="y",ax=ax1,xlim=(0,20),ylim=(0,13))
df2.plot(kind="scatter",x="x",y="y",ax=ax2,xlim=(0,20),ylim=(0,13))
df3.plot(kind="scatter",x="x",y="y",ax=ax3,xlim=(0,20),ylim=(0,13))
df4.plot(kind="scatter",x="x",y="y",ax=ax4,xlim=(0,20),ylim=(0,13))
slope, intercept, r_value, p_value, std_err = stats.linregress(df1["x"],df1["y"])
ax1.plot(df1["x"], intercept + slope*df1["x"], label='fitted line') 
slope, intercept, r_value, p_value, std_err = stats.linregress(df2["x"],df2["y"])
ax2.plot(df2["x"], intercept + slope*df2["x"], label='fitted line') 
slope, intercept, r_value, p_value, std_err = stats.linregress(df3["x"],df3["y"])
ax3.plot(df1["x"], intercept + slope*df3["x"], label='fitted line') 
slope, intercept, r_value, p_value, std_err = stats.linregress(df4["x"],df4["y"])
ax4.plot(df1["x"], intercept + slope*df4["x"], label='fitted line') 
plt.show()  



```
