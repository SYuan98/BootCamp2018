```python
# Problem 1
import pandas as pd
from matplotlib import pyplot as plt
from pydataset import data
iris=data("iris")
iris1=iris.groupby("Species")
iris1.mean()

```

```python
iris1["Petal.Length"].describe()
```

```python
iris1["Petal.Width"].describe()
```

```python
iris1.get_group("setosa").plot(kind="hist", y="Petal.Length", legend="False",
                                                title="setosa Petal.Length")
iris1.get_group("versicolor").plot(kind="hist", y="Petal.Length", legend="False",
                                                title="versicolor Petal.Length")    
iris1.get_group("virginica").plot(kind="hist", y="Petal.Length", legend="False",
                                                title="virginica Petal.Length")    
    
    
```

```python
iris1[["Petal.Length", "Petal.Width"]].mean().plot(kind="barh",
                xerr=iris1.std(), title=r"measurements of iris flowers, $\mu\pm\sigma$")
plt.ylabel("Flowers species")
```

```python
s=data("swiss")
F=pd.qcut(s["Fertility"],8)
s.pivot_table(values="Infant.Mortality",index=F,columns=None,aggfunc="mean")
```

```python
s.cov()
#Catholic is the most important factor for predicting fertility 
```

```python
C=pd.cut(s["Catholic"],[0,20,50,70,90,100])
s.pivot_table(values=["Fertility","Agriculture","Examination","Education","Infant.Mortality"],index=None,columns=C,aggfunc="mean")
```
