```python
# Problem 1
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd 
from scipy import stats
d=np.load("anscombe.npy")
df1=pd.DataFrame(d[:,:2],columns=['x','y'])
df2=pd.DataFrame(d[:,2:4],columns=['x','y'])
df3=pd.DataFrame(d[:,4:6],columns=['x','y'])
df4=pd.DataFrame(d[:,6:],columns=['x','y'])
ax1=plt.subplot(221)
ax2=plt.subplot(222)
ax3=plt.subplot(223)
ax4=plt.subplot(224)
df1.plot(kind="scatter",x="x",y="y",ax=ax1,xlim=(0,20),ylim=(0,13))
df2.plot(kind="scatter",x="x",y="y",ax=ax2,xlim=(0,20),ylim=(0,13))
df3.plot(kind="scatter",x="x",y="y",ax=ax3,xlim=(0,20),ylim=(0,13))
df4.plot(kind="scatter",x="x",y="y",ax=ax4,xlim=(0,20),ylim=(0,13))
slope, intercept, r_value, p_value, std_err = stats.linregress(df1["x"],df1["y"])
ax1.plot(df1["x"], intercept + slope*df1["x"], label='fitted line') 
slope, intercept, r_value, p_value, std_err = stats.linregress(df2["x"],df2["y"])
ax2.plot(df2["x"], intercept + slope*df2["x"], label='fitted line') 
slope, intercept, r_value, p_value, std_err = stats.linregress(df3["x"],df3["y"])
ax3.plot(df1["x"], intercept + slope*df3["x"], label='fitted line') 
slope, intercept, r_value, p_value, std_err = stats.linregress(df4["x"],df4["y"])
ax4.plot(df1["x"], intercept + slope*df4["x"], label='fitted line') 
plt.show()  



```

```python
from scipy import special 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Problem 2

b = lambda n, v, x: special.binom(n, v) * x**v * (1 - x)**(n - v)
x = np.linspace(0, 1)
N = np.arange(0, 10, 1)
fig, axes = plt.subplots(len(N), 
                         len(N), 
                         figsize=(15, 15), 
                         sharex=True, 
                         sharey=True)

for n in N:
    for v in range(n+1):
        axes[n, v].plot(x, b(n, v, x), label=f"n = {n}\nv = {v}")
        axes[n, v].legend(loc='upper right')        
plt.tight_layout()
plt.show()
```

```python
#Problem 4
year, magnitude, longitude, latitude = np.load("earthquakes.npy").T
n = len(year)
plt.plot(year, magnitude)
plt.xlabel("Year")
plt.ylabel("Magnitude")
plt.show()
year_floor = np.floor(year)
year_unique = np.unique(year_floor)
year_freq = np.zeros(len(year_unique))
for i in range(len(year_unique)):
    year_freq[i] = np.sum(year_floor == year_unique[i])
year_freq
plt.bar(year_unique, year_freq)
plt.xlabel('year')
plt.ylabel('Number earthquakes')
plt.title('earthquakes per year')
plt.show()
plt.scatter(latitude, longitude, 10, c=magnitude, cmap = 'coolwarm')
plt.xlabel('latitude')
plt.ylabel('longitude')
cbar = plt.colorbar()
cbar.set_label("magnitude", labelpad = + 1)
plt.show()
```

```python
#problem 5
f = lambda x, y: (1 - x)**2 + 100 * (y - x**2)**2
min_x, max_x = -3, 3
x_grid = np.arange(min_x, max_x, 0.1)
x, y = np.meshgrid(x_grid, x_grid)

plt.pcolormesh(x, y, -f(x, y), cmap='coolwarm')
plt.scatter(1, 1, c='k', alpha=0.7)
plt.show()
```
